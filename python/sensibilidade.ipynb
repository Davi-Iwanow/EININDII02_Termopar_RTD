**Veremos com equações que as formulas da sensibilidade da ponte de Wheatstone é diferente do divisor resistivos para se medir variação de temperatura usando um pt100**

Vamos comparar as fórmulas das sensibilidades da Ponte de Wheatstone e de um divisor resistivo simples para medir variações de temperatura usando um PT100.

**Sensibilidade da Ponte de Wheatstone**:

A sensibilidade da Ponte de Wheatstone é dada por:

$$Sensibilidade_{Wheatstone} = \frac{dV_{out}}{dT}$$

onde $dV_{out}$ é a variação na saída da ponte e $dT$ é a variação na temperatura.

Da fórmula da saída desequilibrada da Ponte de Wheatstone:

$$V_{out} = V_{in} \cdot \frac{\Delta R}{R} $$

e como $\Delta R = \alpha \cdot R \cdot \Delta T$ (variação de resistência do PT100 com a variação de temperatura 
$\Delta T $), substituindo na equação de $V_{out}$:

$$ V_{out} = V_{in} \cdot \frac{\alpha \cdot R \cdot \Delta T}{R} = V_{in} \cdot \alpha \cdot \Delta T$$

A sensibilidade da Ponte de Wheatstone, portanto, é:

$$Sensibilidade_{Wheatstone} = \alpha \cdot V_{in}$$

**Sensibilidade do Divisor Resistivo**:

A sensibilidade do divisor resistivo é dada por:

$$Sensibilidade_{Divisor} = \frac{dV_{out}}{dT}$$

onde $dV_{out}$ é a variação na saída do divisor resistivo e  $dT$ é a variação na temperatura.

Da fórmula da saída do divisor resistivo:

$$V_{out} = V_{in} \cdot \frac{R_2}{R_1 + R_2}$$

A variação $dV_{out}$ devido à variação $dT$ na temperatura pode ser calculada usando a derivada:

$$dV_{out} = \frac{d}{dT} \left( V_{in} \cdot \frac{R_2}{R_1 + R_2} \right) \cdot dT$$

Agora, derivando a expressão em relação a $T$ e substituindo $\Delta R = \alpha \cdot R \cdot \Delta T$ (variação de resistência do PT100 com a variação de temperatura $\Delta T$):

$$dV_{out} = V_{in} \cdot \frac{R_2}{(R_1 + R_2)^2} \cdot \alpha \cdot R \cdot dT$$

A sensibilidade do divisor resistivo é:

$$Sensibilidade_{Divisor} = V_{in} \cdot \frac{R_2}{(R_1 + R_2)^2} \cdot \alpha \cdot R$$

Comparando as duas sensibilidades:

$$Sensibilidade_{Wheatstone} = \alpha \cdot V_{in}$$
$$Sensibilidade_{Divisor} = V_{in} \cdot \frac{R_2}{(R_1 + R_2)^2} \cdot \alpha \cdot R$$

Como podemos ver, as fórmulas das sensibilidades da Ponte de Wheatstone e do divisor resistivo são diferentes. A sensibilidade da Ponte de Wheatstone depende apenas do coeficiente de temperatura $\alpha$ e da tensão de entrada $V_{in}$, enquanto a sensibilidade do divisor resistivo depende adicionalmente dos valores dos resistores $R_1$ e $R_2$, bem como do valor nominal da resistência $R$ do PT100. Isso demonstra que a sensibilidade da Ponte de Wheatstone é mais simples e direta, dependendo apenas das características intrínsecas do sensor de temperatura e da tensão de entrada.
